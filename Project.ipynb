```python
from datetime import datetime
import os
import plotly.io as pio

def save_plot_as_html(fig, filename, insight, start_hour, end_hour):
    global plot_containers

    filepath = os.path.join(html_files_path, filename)

    # Always save the individual HTML file
    fig.write_html(
        filepath,
        full_html=False,
        include_plotlyjs="inline"
    )

    # Current IST hour
    current_hour = datetime.now().hour

    # Only add to dashboard during allowed time
    if start_hour <= current_hour < end_hour:

        html_content = pio.to_html(
            fig,
            full_html=False,
            include_plotlyjs="inline"
        )

        plot_containers += f"""
        <div class="plot-container" id="{filename}" onclick="openPlot('{filename}')">
            <div class="plot">
                {html_content}
            </div>
            <div class="insights">
                {insight}
            </div>
        </div>
        """
    else:
       plot_containers += f"""
       <div class="plot-container">
            <div style="
                display:flex;
                justify-content:center;
                align-items:center;
                height:100%;
                text-align:center;
                font-size:20px;
                font-weight:bold;
                color:white;
                padding:20px;">
                Chart is available only between
                {start_hour}:00 and {end_hour}:00 IST.
            </div>
        </div>
 
        """

    print(filename)
```

```python
save_plot_as_html(fig1, "Category chart 1.html", insight1, 15, 17)
save_plot_as_html(fig2, "Choropleth Map 2.html", insight2, 18, 20)
save_plot_as_html(fig3, "Free vs Paid chart 3.html", insight3, 13, 14)
save_plot_as_html(fig4, "Trend of Installs over time 4.html", insight4, 18, 21)
save_plot_as_html(fig5, "Relationship between App Size and Rating Chart 5.html", insight5, 17, 19)
save_plot_as_html(fig6, "Stacked Area Chart 6.html", insight6, 16, 18)
```

```python
final_html = dashboard_html.format(
    plots=plot_containers,
    plot_width=plot_width,
    plot_height=plot_height
)
```
